```python
from scipy import signal
import matplotlib.pyplot as plt

import random

import numpy as np
```

```python
def wt( data ):
    widths = np.arange(1, 100)
    cwt = signal.cwt( data, signal.ricker, widths )
    fig, ax = plt.subplots( 1, 2, figsize=(20,5) )
    ax[0].plot(data)
    ax[1].imshow(cwt, extent=[-1, 1, 1, 101], cmap='PRGn', aspect='auto', vmax=abs(cwt).max(), vmin=-abs(cwt).max())
```

```python
wt( [random.random()  for i in range(1000)] )
```

```python
wt( [random.random() + np.sin( i / 20 * 3.1415 ) + np.sin( i / 180 * 3.1415 ) for i in range(1000)] )
```
