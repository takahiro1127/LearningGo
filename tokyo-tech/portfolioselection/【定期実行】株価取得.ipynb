```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
%matplotlib inline
```

```python
assetNames = ['AUL','CAN','FRA','GER','JAP', 'USA']
fileNames = ['STW.AX','^GSPTSE' ,'^FCHI','^GDAXI','^N225','^GSPC']
prices = pd.read_csv('price.csv', index_col=0)
```

```python
# Webページを取得して解析する
newDatas = {}
index = 0
for fileName in fileNames:
        load_url = "https://finance.yahoo.com/quote/" + fileName + "?p="+ fileName + "&.tsrc=fin-srch"
        html = requests.get(load_url)
        soup = BeautifulSoup(html.content, "html.parser")
        newDatas[assetNames[index]] = soup.find("span", attrs={'data-reactid': '42', 'class': 'Trsdu(0.3s)'}).get_text().replace(',', '')
        index += 1
```

```python
# csvへ保存
index = 0
for newData in newDatas.values():
        prices.loc[assetNames[index], "2020-08-22"] = newData
        index += 1
prices.to_csv('price.csv')
```
